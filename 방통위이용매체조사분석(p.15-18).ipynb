# PPT 17. OTT 플랫폼 이용자 연령대

```python
#필요없는 u+, 카카오, wa삭제
col=['u+','ka','wa']
user.drop(col,axis=1,inplace=True)
```

```python
#결과 DF 정리하기
ott1 = user.iloc[:,1]
ott2=user.loc[:,['Netflix','wave','Tving','Disney','Coupangplay']]
#넷플릭스, 웨이브, 티빙, 디즈니 쿠팡플레이 이용자 카운팅
ott2['ott_cnt']=ott2.notnull().sum(1)
ott3=user.loc[:,['Q527M_1', 'Q527M_2', 'Q527M_3', 'Q527M_4', 'Q527M_5',
       'Q528M_1', 'Q528M_2', 'Q528M_3', 'Q528M_4', 'Q528M_5']]
result=pd.concat([ott1,ott2,ott3],axis=1)
len(result)
```

```python
ott=result[result['ott_cnt']!=0]
ott.reset_index(inplace=True)
len(ott)
```

# PPT 18. OTT 시청 프로그램 유형

```python
#OTT 시청 프로그램 유형
col=['ott_cnt','Q527M_1','Q527M_2','Q527M_3','Q527M_4','Q527M_5']
ott_type=ott.loc[:,col]
ott_type.reset_index(inplace=True)
len(ott_type)
```

```python
#지상파방송 채널 | 유료방송 채널| OTT 자체 제작 프로그램 |
#이 외 다른 제작형태의 프로그램 | 영화
#채널 유형 이용자 변수 만들기
col=['지상파방송 채널','유료방송 채널','OTT 자체 제작 프로그램','이 외 다른 제작형태의 프로그램','영화']
tot=len(ott_type)
for idx in range(1,6):
  c=f'Q527M_{idx}'
  print(c)
  for i in range(tot):
    if ott_type.at[i,c]==col[4]:
      ott.at[i,'영화']=1

len(ott)
```

```python
#필요없는 컬럼 드랍
dcol=['Q527M_1','Q527M_2','Q527M_3','Q527M_4','Q527M_5']
ott.drop(dcol,axis=1,inplace=True)
```
